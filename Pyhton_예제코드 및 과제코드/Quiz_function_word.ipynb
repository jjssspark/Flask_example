# 함수과제(단어 빈도수 딕셔너리에 담기)

```python
def count_words(*args):
    sentence = input("문장을 입력하세요 : ") # 나는 지금 많이 배고프다
    word_dict = {}
    word_list = sentence.split( )   #["나는", "지금", "많이", "배고프다" ]
    i = 0
    for i in range(0, len(word_list)) :
        count = 0
        for j in range(0, len(word_list)):
            if word_list[i] == word_list[j]:
                count += 1
        word_dict[word_list[i]] = count 
    return word_dict

final = count_words()
total = sum(final.values())
print(final)
print("총 단어 수 : ", total)

                
            
    
    
    
    
```
