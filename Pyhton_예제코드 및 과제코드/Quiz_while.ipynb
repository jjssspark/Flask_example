# while 사용 퀴즈


```python
while(True):
    num1 = input("첫 번째 숫자를 입력하시오 : ")
    if num1 == "q":
        break
    num2 = input("두 번째 숫자를 입력하시오 : ")
    if num2 == "q":
        break
    opr = input("연산자를 입력하시오 : ")
    if opr == "q":
        break
    
    if opr == "+":
        print(f"{num1} {opr} {num2} = ", int(num1) + int(num2))
        continue
    elif opr == "-":
        print(f"{num1} {opr} {num2} = ",int(num1) - int(num2))
        continue
    elif opr == "*":
        print(f"{num1} {opr} {num2} = ", int(num1) * int(num2))
        continue
    elif opr == "/":
        if num2 == "0" :
            print("0으로 나눌 수 없습니다")
        else:
            print(f"{num1} {opr} {num2} = ", int(num1) / int(num2))
        continue
    else:
        print("잘못된 연산자입니다")
        continue
```
