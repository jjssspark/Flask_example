# 클래스(Class)

```python
# 클래스 속성 및 인스턴스 속성 예시
class Dog:   # 클래스 생성 / 강아지 설계도 만들기
    # 클래스 속성 : 모든 개가 공유
    species = "Puddle" # 모든 강아지가 공통으로 갖는 값 -> 모든 강아지는 같은 설계도를 공유함
    
    def __init__(self, dog_name, dog_age): # 생성자 -> 강아지가 만들어질때 이 함수가 자동으로 실행됨
        # 인스턴스 속성 : 각 개별 개에 독립적으로 저장되는 정보들
        self.name = dog_name
        self.age = dog_age

# 사용 예시
dog1 = Dog("멍멍이", 3)
dog2 = Dog("흰둥이", 5)

print(dog1.species)
print(dog2.species)
print(dog1.name)
print(dog2.name)
print(dog1.age)
print(dog2.age)
```

```python
# 메서드(Methods) 예시
class Dog:
    species = "Puddle"  # 클래스 속성
    
    def __init__(self, name, age):
        self.name = name  # 인스턴스 속성
        self.age = age  # 인스턴스 속성
        
    # 인스턴스 메서드 : 인스턴스 속성 사용
    def bark(self):
        return f"{self.name}가 짖습니다! (나이 : {self.age})"
    
    #클래스 메서드 : 클래스 속성 사용
    @classmethod  # cls : 클래스 자체
    def get_species(cls):
        return cls.species
    
    # 정적 메서드 : 독립적 작업(클래스 / 인스턴스와 무관)
    @staticmethod
    def make_sound():
        return "멍멍!"
    
# 사용 예시
dog = Dog("멍멍이",3)
    
print(dog.bark())
print(Dog.get_species())
print(Dog.make_sound())
```

```python
class Dog:
    # 초기화 메서드(생성자)
    def __init__(self, name):
        self.name = name # 인스턴스 변수
    
    def bark(self): # 인스턴스 메서드
        print(f"{self.name}가 짖습니다!")
# 객체 생성
my_dog = Dog("멍멍이")
my_dog.bark()
```

```python
# 상속 예제
# 부모 클래스 정의
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        pass
# 자식 클래스 정의
class Cat(Animal): # Animal 클래스 상속
    def speak(self):
        return f"{self.name}가 야옹하고 웁니다"
    
# 객체 생성 및 사용
cat1 = Cat("나비") # Cat 클래스의 인스턴스 생성
print(cat1.name)
print(cat1.speak())

cat2 = Cat("미야")
print(cat2.name)
print(cat2.speak())
```

```python
# 캡슐화 예제
class BankAccount:
    def __init__(self):
        self.__balance = 0 # 비공개 변수
        
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    def get_balance(self):
        return self.__balance

# 실행 예제
account = BankAccount()
account.deposit(1000)
print(account.get_balance())
#print(account.__balance)
```

```python
class Wallet:
    def __init__(self, money):
        self.__money = money # 돈을 비공개로 저장
    def check_money(self):
        return self.__money # 돈을 확인하는 방법 제공
    def add_money(self, amount):
        self.__money += amount # 돈을 추가하는 방법 제공

# 지갑 만들기
my_wallet = Wallet(10000)

# 돈 확인
print(my_wallet.check_money())

# 돈 추가
my_wallet.add_money(5000)
print(my_wallet.check_money())

#직접 접근 시도
#print(my_wallet.__money) #에러! AttributeError: 'Wallet' object has no attribute '__money'
```

```python
# 클래스 변수와 인스턴스(객체) 변수 예제
class Student:   # 학생 설계도
    school_name = "파이썬 고등학교"  # 클래스 변수 = 모든 학생이 공유하는 변수
    student_count = 0  # 클래스 변수 = 모든 학생이 공유하는 변수
    
    def __init__(self, name):  # 생성자 -> 학생 만들어질때 실행함
        self.name = name  # 인스턴스 변수   -> 학생마다 다른 정보
        Student.student_count += 1   # 학생이 생성될 때마다 학생 수의 증가
# 실행 예제
student1 = Student("김철수")
student2 = Student("이영희")

print(Student.school_name)
print(student1.school_name)
print(student2.school_name)
print(Student.student_count)
```

```python
# 간단한 계산기 클래스
class Calculator:
    # 초기화 메서드 : 아무 상태 없이 계산만 수행
    def __init__(self):  # __init__ : 클래스의 인스턴스가 호출될 때 자동으로 호출되는 메서드
        print("계산기를 초기화합니다.")
    
    # 두 숫자를 더하는 메서드
    def add(self, a, b):
        return a + b
    
    # 두 숫자를 곱하는 메서드
    def multiply(self, a, b):
        return a * b

# 객체 생성 및 메서드 호출
calc = Calculator()
print("덧셈 결과 : ", calc.add(10,20))
print("곱셈 결과 : ", calc.multiply(10,20))
```

```python
# 학생정보관리 클래스 예제
class Student:
    # 생성자 : 학생 이름과 나이를 초기화
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    # 학생 정보 출력 메서드
    def display_info(self):
        print(f"이름 : {self.name}, 나이 : {self.age}")
     
# 객체 생성 및 메서드 호출
student1 = Student("홍길동", 16)
student2 = Student("박지수", 25)
    
student1.display_info()
student2.display_info()
```

```python
# 영화 티켓 할인 예제
class MovieTicket:
    # 생성자 : 영화 제목과 기분 티켓 가격 초기화
    def __init__(self, title, price):
        self.title = title
        self.price = price
    
    # 할인 적용 메서드
    def apply_discount(self, discount):
        self.price -= discount
        print(f"{self.title} 영화의 할인된 가격 : {self.price}원")
        

# 객체 생성 및 메서드 호출
ticket = MovieTicket("어벤져스", 12000)
ticket.apply_discount(2000)
```

```python
# 간단한 은행 계좌 클래스
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount}원이 입금되었습니다. 현재 잔액 : {self.balance}원")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
        else:
            self.balance -= amount
            print(f"{amount}원이 출금되었습니다. 현재 잔액 : {self.balance}원")

# 사용 예시
account = BankAccount("홍길동", 10000)
account.deposit(5000)
account.withdraw(2000)
```
