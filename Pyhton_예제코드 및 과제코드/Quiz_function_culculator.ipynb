# 함수 과제(사칙연산 계산기)


```python
def calculate(num1, num2, operator):
    if operator == "+":
        result = num1 + num2
    elif operator == "-":
        result = num1 - num2
    elif operator == "*":
        result = num1 * num2
    elif operator == "/":
        if num2 == 0:
            result = "Division by zero"
        else:
            result = num1 / num2
    else:
        result = "Invalid operator"
    
    return result

num1 = int(input("첫 번째 숫자를 입력하세요 : "))
num2 = int(input("두 번째 숫자를 입력하세요 : "))
operator = input("연산자를 선택하세요(+, - , * , /) : ")

print("결과 -> ", calculate(num1, num2, operator))
```
