```python
#[start:end:step] 형식
text = "Python"
print(text[0:3])
print(text[::2]) # 2칸씩 건너뜀
```

```python
# upper() : 모든 문자를 대문자로 변환
# lower() : 모든 문자를 소문자로 변환
text = "Python"
print(text.upper())
print(text.lower())
```

```python
# find() : 특정 문자의 인덱스 반환(없으면 -1 반환)
# count() : 특정 문자의 개수 세기
text = "banana"
print(text.find('a'))
print(text.count("a"))
```

```python
# strip() : 문자열 양쪽 끝 공백 제거
# lstrip() rstrip() : 왼쪽 또는 오른쪽 공백만 제거
text = " Hello World "
print(text)
print(text.strip())
```

```python
# split() : 특정 구분자로 문자열을 나누어 리스트로 반환
# join() : 리스트의 문자열을 특정 구분자로 연결하여 하나의 문자열로 반환
text = "Python is fun"
words = text.split()
print(words)
joined_text = " ".join(words)
print(joined_text)
```

```python
# 예시 : 파이썬 3.6이상 f - string 방식
name = "Tina"
age = 25
print(f"My name is {name} and I am {age} years old")
```

```python
print("Hello\nPython") # \n : 줄바꿈
print("탭을\t사용한\t예시") # \t : 탭
print("백슬래시는 이렇게 \\출력") # \\ : 백슬래시 
```

#리스트

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0])
```

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0:2])
```

```python
numbers = [3, 1, 4, 1, 5]
numbers.append(9) # append : 요소 추가 함수
print(numbers)
```

```python
numbers.sort() # sort : 요소 정렬 함수
print(numbers)
```

```python
numbers.remove(1) # remove : 요소 제거 함수
print(numbers)
```

```python
removed_item = numbers.pop(2) # pop : 특정 인덱스 위치의 요소 추출 및 삭제
print(numbers)
print(removed_item)
```

```python
numbers.reverse() # reverse : 리스트 뒤집기
print(numbers)
```

```python
count = numbers.count(5) # count : 리스트 요소의 개수 세기
print(count)
```

```python
new_numbers = numbers.copy()
print(new_numbers)
```

```python
numbers.extend([6, 7, 8]) # extend : 리스트 확장
print(numbers)
```

```python
another_list = [10, 11, 12]
combined_list = numbers + another_list # 각 리스트를 + 로 여결
print(combined_list)
```
